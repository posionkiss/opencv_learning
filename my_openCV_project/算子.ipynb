识别并留住图像边缘

边缘是像素值发生跃迁的位置，是图像的显著特征之一，在图像特征提取，对象检测，模式识别等方面都有重要的作用。

索贝尔sobel算子
sobel算子对图像求一阶导数。一阶导数越大，说明像素在该方向的变化越大，边缘信号越强。因为图像的灰度值都是离散的数字，sobel算子采用离散差分算子计算图像像素点亮度值的近似梯度

图像是二维的，即沿着宽度/高度两个方向，我们使用两个卷积核对图像进行处理：


它能够近似计算出一阶导数
综合考虑这两个方向的变化，我们使用公式反映某个像素的梯度变化情况：G=GradiantX和Y平方和开根号, 或绝对值之和


```python
import cv2
import numpy as np
chess=cv2.imread('../material/used/chess.png')
cv2.imshow('chess',chess)
chess_sobx=cv2.Sobel(chess,-1,1,0,ksize=3)#-1表示返回和chess一样的数据类型，1,0表示只对x方向求一阶导
chess_soby=cv2.Sobel(chess,-1,0,1)
#chess_sob=chess_sobx+chess_soby
chess_sob=cv2.add(chess_sobx,chess_soby)
cv2.imshow('chess_sob',chess_sob)


while cv2.waitKey(0)!=ord('q'):
    None
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
pic=cv2.imread('../material/used/wall pic16.png')
pic=cv2.resize(pic,dsize=None,fx=0.16,fy=0.16)
cv2.imshow('pic',pic)
pic_schx=cv2.Scharr(pic,-1,1,0)
pic_schy=cv2.Scharr(pic,-1,0,1)
cv2.imshow('pic_schx',pic_schx)
cv2.imshow('pic_schy',pic_schy)

pic_sch=cv2.add(pic_schx,pic_schy)
cv2.imshow('pic_sch',pic_sch)


while cv2.waitKey(0)!=ord('q'):
    None
cv2.destroyAllWindows()
```
